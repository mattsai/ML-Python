```python
reviews.loc[0,'country'] #If I have index that is not 0, will be .loc['IindexName','country']
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']] #All data only those columns
reviews.index #we can see that index are int
```

```python
reviews.set_index('title') #we set the column title as a index
reviews.country == 'Italy' #will get a series if the country is or not italy booleans True/False
reviews.loc[reviews.country == 'Italy'] #Will get only the True values

#Now we want when the country is italy and points are more than 90
#you will need to close () and use the & ampersand symbol
reviews.loc[(reviews.country=='Italy') & (reviews.points>=90)]
```

```python
#isin is like "and of vlaues" is in this and is in here and etc
reviews.loc[reviews.country.isin(['Italy'])] #what is the same as reviews.country == 'Italy'
reviews.loc[reviews.country.isin(['Italy','France'])] #Here is when it works better rather than having two reviews.country == Italy and reviews.country == France statements
# reviews.loc[]
```

```python
#isnull notnull
reviews.loc[reviews.points.isnull()]
reviews.loc[reviews.points.notnull()]
```
